```python
start = time.time()
reviews = list()

for i in range(7):
    filename = 'review_words-' + str(i+1) + '.pkl'
    print('Opening', filename, end='  ')
    with open(filename, 'rb') as f:
        reviews_set = pickle.load(f)
        print('Contains', len(reviews_set), 'entries')
        reviews.extend(reviews_set)

end = time.time()
print('Loading data', 'Time taken:', end - start)
len(reviews)
```

```python
start = time.time()
word_count = dict()
SLICE = 10000

for review in reviews[:SLICE]:
    for sentence in review:
        for word in sentence:
            if not word in word_count:
                word_count[word] = 1
            else:
                word_count[word] += 1
                
vocab_size = len(word_count)
end = time.time()
print('Word count', 'Time taken:', end - start)
vocab_size
```

```python
rare_count = 0

for word in word_count.keys():
    if word_count[word] < 5:
        rare_count += 1
        print(word)
```

```python
print(rare_count)
```

```python
word_list = list(word_count.keys())
len(word_list)
```

```python
start = time.time()
co_occ = np.zeros((vocab_size, vocab_size))

for i, review in enumerate(reviews[:SLICE]):
    if i % 1000 == 0:
        print(i)
    for sentence in review:
        sent_len = len(sentence)
        for wi in range(sent_len):
            curr_word = sentence[wi]
            curr_ind = word_list.index(curr_word)

            left_ind = max(0, wi-3)
            right_ind = min(sent_len, wi+4)

            for cont_word_ind in range(left_ind, right_ind):
                cont_word = sentence[cont_word_ind]
                cont_ind = word_list.index(cont_word)
                co_occ[curr_ind, cont_ind] += 1

end = time.time()
print('Time taken:', end - start)
co_occ.shape
```

```python
start = time.time()
sparsity = (1 - np.count_nonzero(co_occ) / co_occ.size) * 100
print('Sparsity:', sparsity)

co_occ_csr = sp.csr_matrix(co_occ)
end = time.time()

print('Time taken:', end - start)
co_occ_csr
```

```python
start = time.time()
u, s, vt = sp.linalg.svds(co_occ_csr, k=750, which='LM')
end = time.time()

print('Time taken:', end - start)
print(u.shape, s.shape, vt.shape)
```
